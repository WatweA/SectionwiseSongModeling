```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corpus_df = pd.read_csv('cleaned_section_headers/all_lyrics_cleaned.csv')
corpus_df
```

```python
genres = ['country','hiphop','pop','rock']
def max_one(num):
    return min(num, 1)
for genre in genres:
    corpus_df[genre] = corpus_df[genre].map(max_one)
    
corpus_df
```

```python
song_genre_heatmap = sns.heatmap(corpus_df.corr(), cmap="Blues_r", annot=True, vmin=-1, vmax=1)
song_genre_heatmap.figure.savefig('song_genre_heatmap.png', dpi=300, bbox_inches='tight')
song_genre_heatmap
```
